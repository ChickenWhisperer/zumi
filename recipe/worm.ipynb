```python
import cv2
import IPython.display
import PIL.Image

# Load a photo of Donald Trump
chicken = cv2.imread('images/chicken.jpg')

# Convert the image space from BGR (Blue, Green, Red) to RGB
chicken = cv2.cvtColor(chicken, cv2.COLOR_BGR2RGB)

# Resize Trump
chicken = cv2.resize(chicken,(800,450))
# Draw a diagonal blue line with thickness of 5 px
cv2.ellipse(chicken,(300,100),(20,10),0,0,360,(250,225,0),-1)
cv2.ellipse(chicken,(310,100),(20,10),0,0,360,(225,200,0),-1)
cv2.ellipse(chicken,(330,100),(20,10),0,0,360,(250,225,0),-1)
cv2.ellipse(chicken,(340,100),(20,10),0,0,360,(225,200,0),-1)
cv2.ellipse(chicken,(350,100),(20,10),0,0,360,(250,225,0),-1)
cv2.ellipse(chicken,(360,100),(20,10),0,0,360,(225,200,0),-1)
# Show the image of Trump
IPython.display.display(PIL.Image.fromarray(chicken))

# CHALLENGE: Use the openCV documentation to draw red eyes on Donald Trump
# https://docs.opencv.org/3.1.0/dc/da5/tutorial_py_drawing_functions.html
```
